```python
# Phase one - situation analysis
# paharmaLens 


# importing libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the DS
dataset_path="E:\PharmaLens\data\Pharma_BI_Dataset_v3.xlsx"



# i just added the header cause the first row is the title of the sheet, so I just needed to tell that the names of the columns are in row 2 
sales = pd.read_excel(dataset_path , sheet_name="Sales Transactions" , header=1)

expenses = pd.read_excel(dataset_path , sheet_name="Expenses" , header=1)
dept = pd.read_excel(dataset_path , sheet_name="Debt & AR" ,  header=1)
inventory = pd.read_excel(dataset_path , sheet_name="Inventory (HQ)",header=1)




print("Data Loaded sucessfully!!")
print(f"sales rows : {len(sales):,}")
print(f"Expenses rows : {len(expenses):,}")
print(f"Dept rows : {len(dept):,}")
print(f"Inventory rows : {len(inventory):,}")





```

```python
# overall company performance   >>  the thing is, we need  to know the entire  performance of the company 
#  we know total revenue,  total cost, total profit, and so on  

total_revenue = sales["Total Revenue EGP"].sum()
total_cost= sales["Total Cost EGP"].sum()
total_profit=sales["Gross Profit EGP"].sum()
profit_marign=(total_profit/total_revenue)*100

print("=" * 40 )
print("PHARMALENS — COMPANY OVERVIEW")
print("=" * 40 )
print(f"total revenue : {total_revenue: ,.2f}")
print(f"total cost : {total_cost: ,.2f}")
print(f"total profit : {total_profit: ,.2f}")
print(f"profit marignt : {profit_marign: ,.2f}%")
print("=" * 40 )














 


```

```python
# so the good news is for every 100 EGP we sell we keep >>  43 EGP, I think that is a good margin 


# We need to break this down by year to know if we are growing or shrinking  



# so this gross profit is so healthy now, but these are gross profits — before expenses like salaries, rent, logistics
# So we need to find out how much expenses eat into this profit


# Revenue comparison 2023 vs 2024
yearly = sales.groupby("Year").agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum")
).reset_index()

yearly["Profit_Margin_%"] = (yearly["Total_Profit"] / yearly["Total_Revenue"]) * 100





print("=" * 50)
print("   PHARMALENS — 2023 vs 2024 COMPARISON")
print("=" * 50)
for _, row in yearly.iterrows():
    print(f"\n  Year: {int(row['Year'])}")
    print(f"     Revenue : EGP {row['Total_Revenue']:,.2f}")
    print(f"     Cost    : EGP {row['Total_Cost']:,.2f}")
    print(f"     Profit  : EGP {row['Total_Profit']:,.2f}")
    print(f"     Margin  : {row['Profit_Margin_%']:.2f}%")
print("\n" + "=" * 50)




    









```

```python
# The good news is that the company is  growing 
# Revenue has increased by 264,728.24   
# profit also increased 115,674.59 
# margin is almost identical, which is good  - cost control is stable

# The cost is also rising, so we need to make sure the cost does not  grow faster than the revenue 


#  so i  will continue digging to know where the profit is coming from and where the money is being lost 






```

```python
# i will break down to know which branch is driving this growth 




gov_performance = sales.groupby(["Governorate", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum")
).reset_index()



gov_performance["Margin_%"] = (gov_performance["Total_Profit"] / gov_performance["Total_Revenue"]) * 100





print("=" * 60)
print("   PHARMALENS — PERFORMANCE BY GOVERNORATE")
print("=" * 60)
for gov in gov_performance["Governorate"].unique():
    print(f"\n {gov}")
    gov_data = gov_performance[gov_performance["Governorate"] == gov]
    for _, row in gov_data.iterrows():
        print(f"     {int(row['Year'])} → Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 60)
     
```

```python
"""
cairo  is the star for cause, also kafr and  baheria are growing  

  in gharbia is almost flat. The reps need to push harder there 

 Mansoura is the only branch that loses  revenue, so it may be the market there  or the reps

 cairo should get more  investment  and attention 

"""





```

```python
# so nex i want to know which of the four medicines makes the most money '



# Revenue by Medicine
medicine_performance = sales.groupby(["Medicine Name", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

medicine_performance["Margin_%"] = (medicine_performance["Total_Profit"] / medicine_performance["Total_Revenue"]) * 100

print("=" * 65)
print("   PHARMALENS — PERFORMANCE BY MEDICINE")
print("=" * 65)
for med in medicine_performance["Medicine Name"].unique():
    print(f"\n  {med}")
    med_data = medicine_performance[medicine_performance["Medicine Name"] == med]
    for _, row in med_data.iterrows():
        print(f"     {int(row['Year'])} → Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 65)
```

```python

"""
Atorvastatin is your biggest money maker — EGP 534K profit in 2024 alone
     Highest margin at 43.59% and growing in quantity every year
     This is the medicine to push hardest



   Vitamin D3 is growing the fastest in quantity — +1,800 units
   Second-highest profit and margin — very strong performer
   Demand is clearly increasing year over year

"""
```

```python
# so i need to know how the medicines is doing in each gov which makes the most money and  which makes the lowest money 


gov_med = sales.groupby(["Governorate", "Medicine Name"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

gov_med["Margin_%"] = (gov_med["Total_Profit"] / gov_med["Total_Revenue"]) * 100

print("=" * 65)
print("   PHARMALENS — MEDICINE PERFORMANCE PER GOVERNORATE")
print("=" * 65)
for gov in gov_med["Governorate"].unique():
    print(f"\n   {gov}")
    gov_data = gov_med[gov_med["Governorate"] == gov].sort_values("Total_Profit", ascending=False)
    for _, row in gov_data.iterrows():
        print(f"     {row['Medicine Name']:<22} → Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 65)



```

```python

"""
Atorvastatin is #1 in EVERY single governorate without exception
Vitamin D3 is #2 in EVERY single governorate without exception
This is very clear — these two medicines are your company's backbone

 Cairo Dominates:

Cairo generates the highest profit from every single medicine
Atorvastatin alone in Cairo makes EGP 293K — almost double Beheira
Cairo has the biggest market and your reps there should push even harder

 Panadol is Last Everywhere:

Panadol is the least profitable medicine in every governorate
Lowest margin AND lowest profit across the board
Not worth heavy investment compared to Atorvastatin

Mansoura Warning:

Even though Atorvastatin is #1 in Mansoura
The overall numbers are lower than expected for its size
Confirms our earlier finding — Mansoura is underperforming


The Big Insight:

Your reps in every governorate should be trained to push Atorvastatin and Vitamin D3 first. These two medicines alone account for the majority of your profits everywhere
"""
```

```python
# now I move to the analysis of the employees 


# Performance by Medical Rep
rep_performance = sales.groupby(["Rep Name", "Governorate", "Area"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum"),
    Total_Invoices=("Invoice ID", "count")
).reset_index()

rep_performance["Margin_%"] = (rep_performance["Total_Profit"] / rep_performance["Total_Revenue"]) * 100
rep_performance = rep_performance.sort_values("Total_Profit", ascending=False)

print("=" * 70)
print("   PHARMALENS — MEDICAL REP PERFORMANCE (2023-2024)")
print("=" * 70)
for _, row in rep_performance.iterrows():
    print(f"\n   {row['Rep Name']:<20} | {row['Governorate']:<15} | Area: {row['Area']}")
    print(f"     Invoices: {int(row['Total_Invoices']):,} | Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f}")
print("\n" + "=" * 70)
```

```python

collector_performance = dept.groupby(["Collector Name", "Governorate"]).agg(
    
    Total_Invoices=("Debt ID", "count"),
    Total_Debt=("Invoice Amount EGP", "sum"),
    Total_Collected=("Collected EGP", "sum"),
    Total_Outstanding=("Outstanding EGP", "sum")
).reset_index()

collector_performance["Collection_Rate_%"] = (collector_performance["Total_Collected"] / collector_performance["Total_Debt"]) * 100
collector_performance = collector_performance.sort_values("Collection_Rate_%", ascending=False)

print("=" * 70)
print("   PHARMALENS — COLLECTOR PERFORMANCE (2023-2024)")
print("=" * 70)
rank = 1
for _, row in collector_performance.iterrows():
    print(f"\n  #{rank}  {row['Collector Name']:<20} | {row['Governorate']}")
    print(f"      Invoices: {int(row['Total_Invoices']):,} | Total Debt: EGP {row['Total_Debt']:,.2f} | Collected: EGP {row['Total_Collected']:,.2f} | Outstanding: EGP {row['Total_Outstanding']:,.2f} | Collection Rate: {row['Collection_Rate_%']:.2f}%")
    rank += 1
print("\n" + "=" * 70)
```
