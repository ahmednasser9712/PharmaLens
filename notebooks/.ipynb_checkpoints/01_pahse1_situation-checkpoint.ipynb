```python
#load the DS
dataset_path="E:\PharmaLens\data\Pharma_BI_Dataset_v3.xlsx"



# i just added the header cause the first row is the title of the sheet, so I just needed to tell that the names of the columns are in row 2 
sales = pd.read_excel(dataset_path , sheet_name="Sales Transactions" , header=1)

expenses = pd.read_excel(dataset_path , sheet_name="Expenses" , header=1)
dept = pd.read_excel(dataset_path , sheet_name="Debt & AR" ,  header=1)
inventory = pd.read_excel(dataset_path , sheet_name="Inventory (HQ)",header=1)




print("Data Loaded sucessfully!!")
print(f"sales rows : {len(sales):,}")
print(f"Expenses rows : {len(expenses):,}")
print(f"Dept rows : {len(dept):,}")
print(f"Inventory rows : {len(inventory):,}")





```

```python
# overall company performance   >>  the thing is, we need  to know the entire  performance of the company 
#  we know total revenue,  total cost, total profit, and so on  

total_revenue = sales["Total Revenue EGP"].sum()
total_cost= sales["Total Cost EGP"].sum()
total_profit=sales["Gross Profit EGP"].sum()
profit_marign=(total_profit/total_revenue)*100

print("=" * 40 )
print("PHARMALENS — COMPANY OVERVIEW")
print("=" * 40 )
print(f"total revenue : {total_revenue: ,.2f}")
print(f"total cost : {total_cost: ,.2f}")
print(f"total profit : {total_profit: ,.2f}")
print(f"profit marignt : {profit_marign: ,.2f}%")
print("=" * 40 )














 


```

```python
# so the good news is for every 100 EGP we sell we keep >>  43 EGP, I think that is a good margin 


# We need to break this down by year to know if we are growing or shrinking  



# so this gross profit is so healthy now, but these are gross profits — before expenses like salaries, rent, logistics
# So we need to find out how much expenses eat into this profit


# Revenue comparison 2023 vs 2024
yearly = sales.groupby("Year").agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum")
).reset_index()

yearly["Profit_Margin_%"] = (yearly["Total_Profit"] / yearly["Total_Revenue"]) * 100





print("=" * 50)
print("   PHARMALENS — 2023 vs 2024 COMPARISON")
print("=" * 50)
for _, row in yearly.iterrows():
    print(f"\n  Year: {int(row['Year'])}")
    print(f"     Revenue : EGP {row['Total_Revenue']:,.2f}")
    print(f"     Cost    : EGP {row['Total_Cost']:,.2f}")
    print(f"     Profit  : EGP {row['Total_Profit']:,.2f}")
    print(f"     Margin  : {row['Profit_Margin_%']:.2f}%")
print("\n" + "=" * 50)




    









```

```python
# The good news is that the company is  growing 
# Revenue has increased by 264,728.24   
# profit also increased 115,674.59 
# margin is almost identical, which is good  - cost control is stable

# The cost is also rising, so we need to make sure the cost does not  grow faster than the revenue 


#  so i  will continue digging to know where the profit is coming from and where the money is being lost 






```

```python
# i will break down to know which branch is driving this growth 




gov_performance = sales.groupby(["Governorate", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum")
).reset_index()



gov_performance["Margin_%"] = (gov_performance["Total_Profit"] / gov_performance["Total_Revenue"]) * 100





print("=" * 60)
print("   PHARMALENS — PERFORMANCE BY GOVERNORATE")
print("=" * 60)
for gov in gov_performance["Governorate"].unique():
    print(f"\n {gov}")
    gov_data = gov_performance[gov_performance["Governorate"] == gov]
    for _, row in gov_data.iterrows():
        print(f"     {int(row['Year'])} → Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 60)
     
```

```python
# so nex i want to know which of the four medicines makes the most money '



# Revenue by Medicine
medicine_performance = sales.groupby(["Medicine Name", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Cost=("Total Cost EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

medicine_performance["Margin_%"] = (medicine_performance["Total_Profit"] / medicine_performance["Total_Revenue"]) * 100

print("=" * 65)
print("   PHARMALENS — PERFORMANCE BY MEDICINE")
print("=" * 65)
for med in medicine_performance["Medicine Name"].unique():
    print(f"\n  {med}")
    med_data = medicine_performance[medicine_performance["Medicine Name"] == med]
    for _, row in med_data.iterrows():
        print(f"     {int(row['Year'])} → Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 65)
```

```python
# so i need to know how the medicines is doing in each gov which makes the most money and  which makes the lowest money 


gov_med = sales.groupby(["Governorate", "Medicine Name"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

gov_med["Margin_%"] = (gov_med["Total_Profit"] / gov_med["Total_Revenue"]) * 100

print("=" * 65)
print("   PHARMALENS — MEDICINE PERFORMANCE PER GOVERNORATE")
print("=" * 65)
for gov in gov_med["Governorate"].unique():
    print(f"\n   {gov}")
    gov_data = gov_med[gov_med["Governorate"] == gov].sort_values("Total_Profit", ascending=False)
    for _, row in gov_data.iterrows():
        print(f"     {row['Medicine Name']:<22} → Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f} | Margin: {row['Margin_%']:.2f}%")
print("\n" + "=" * 65)



```

```python
# now I move to the analysis of the employees 


# Performance by Medical Rep
rep_performance = sales.groupby(["Rep Name", "Governorate", "Area"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum"),
    Total_Invoices=("Invoice ID", "count")
).reset_index()

rep_performance["Margin_%"] = (rep_performance["Total_Profit"] / rep_performance["Total_Revenue"]) * 100
rep_performance = rep_performance.sort_values("Total_Profit", ascending=False)

print("=" * 70)
print("   PHARMALENS — MEDICAL REP PERFORMANCE (2023-2024)")
print("=" * 70)
for _, row in rep_performance.iterrows():
    print(f"\n   {row['Rep Name']:<20} | {row['Governorate']:<15} | Area: {row['Area']}")
    print(f"     Invoices: {int(row['Total_Invoices']):,} | Qty: {int(row['Total_Quantity']):,} | Revenue: EGP {row['Total_Revenue']:,.2f} | Profit: EGP {row['Total_Profit']:,.2f}")
print("\n" + "=" * 70)
```

```python

collector_performance = dept.groupby(["Collector Name", "Governorate"]).agg(
    
    Total_Invoices=("Debt ID", "count"),
    Total_Debt=("Invoice Amount EGP", "sum"),
    Total_Collected=("Collected EGP", "sum"),
    Total_Outstanding=("Outstanding EGP", "sum")
).reset_index()

collector_performance["Collection_Rate_%"] = (collector_performance["Total_Collected"] / collector_performance["Total_Debt"]) * 100
collector_performance = collector_performance.sort_values("Collection_Rate_%", ascending=False)

print("=" * 70)
print("   PHARMALENS — COLLECTOR PERFORMANCE (2023-2024)")
print("=" * 70)
rank = 1
for _, row in collector_performance.iterrows():
    print(f"\n  #{rank}  {row['Collector Name']:<20} | {row['Governorate']}")
    print(f"      Invoices: {int(row['Total_Invoices']):,} | Total Debt: EGP {row['Total_Debt']:,.2f} | Collected: EGP {row['Total_Collected']:,.2f} | Outstanding: EGP {row['Total_Outstanding']:,.2f} | Collection Rate: {row['Collection_Rate_%']:.2f}%")
    rank += 1
print("\n" + "=" * 70)
```

```python
# The thing we need to do is to know where the money is leaking
# I am going to analyze the expenses  to know  if there is any leakage or not 



exp_performance = expenses.groupby(["Governorate", "Category"]).agg(
    Total_Budget=("Budget EGP", "sum"),
    Total_Actual=("Actual EGP", "sum"),
    Total_Variance=("Variance EGP", "sum")
).reset_index()


exp_performance["Variance_%"] = (exp_performance["Total_Variance"] / exp_performance["Total_Budget"]) * 100
exp_performance = exp_performance.sort_values("Total_Variance", ascending=False)



print("=" * 70)
print("   PHARMALENS — EXPENSE ANALYSIS (2023-2024)")
print("=" * 70)



for gov in exp_performance["Governorate"].unique():
    print(f"\n   {gov}")
    gov_data = exp_performance[exp_performance["Governorate"] == gov].sort_values("Total_Variance", ascending=False)
    total_budget = gov_data["Total_Budget"].sum()
    total_actual = gov_data["Total_Actual"].sum()
    total_variance = gov_data["Total_Variance"].sum()
    print(f"     Total Budget: EGP {total_budget:,.2f} | Total Actual: EGP {total_actual:,.2f} | Variance: EGP {total_variance:,.2f}")
    print(f"     {'Category':<20} {'Budget':>15} {'Actual':>15} {'Variance':>15} {'Variance%':>10}")
    print(f"     {'-'*75}")
    for _, row in gov_data.iterrows():
        flag = "⚠️" if row["Variance_%"] > 15 else "✅"
        print(f"     {flag} {row['Category']:<18} {row['Total_Budget']:>15,.2f} {row['Total_Actual']:>15,.2f} {row['Total_Variance']:>15,.2f} {row['Variance_%']:>9.2f}%")
print("\n" + "=" * 70)




    

```

```python
"""
The big insight is that the company is overspending by 598k   across all branches - if  we control just salaries and miscellaneous, we 
Can like save about 200k  and turn it directly into profit 
"""
```

```python
#  i need to kanow who is overdue like who owe us the most money 






debt_analysis = dept.groupby(["Governorate", "Customer Name", "Risk Level"]).agg(
    Total_Invoices=("Debt ID", "count"),
    Total_Debt=("Invoice Amount EGP", "sum"),
    Total_Collected=("Collected EGP", "sum"),
    Total_Outstanding=("Outstanding EGP", "sum"),
    Avg_Days_Overdue=("Days Overdue", "mean")
).reset_index()

debt_analysis["Collection_Rate_%"] = (debt_analysis["Total_Collected"] / debt_analysis["Total_Debt"]) * 100
debt_analysis = debt_analysis.sort_values("Total_Outstanding", ascending=False)

print("=" * 75)
print("   PHARMALENS — DEBT & AR ANALYSIS (2024)")
print("=" * 75)


total_debt = debt_analysis["Total_Debt"].sum()
total_collected = debt_analysis["Total_Collected"].sum()
total_outstanding = debt_analysis["Total_Outstanding"].sum()
overall_collection = (total_collected / total_debt) * 100

print(f"\n   OVERALL SUMMARY:")
print(f"     Total Debt      : EGP {total_debt:,.2f}")
print(f"     Total Collected : EGP {total_collected:,.2f}")
print(f"     Outstanding     : EGP {total_outstanding:,.2f}")
print(f"     Collection Rate : {overall_collection:.2f}%")

# High risk customers
print(f"\n   HIGH RISK CUSTOMERS:")
print(f"  {'Customer':<30} {'Gov':<15} {'Outstanding':>15} {'Days Overdue':>12} {'Collection%':>12}")
print(f"  {'-'*85}")
high_risk = debt_analysis[debt_analysis["Risk Level"] == "High"].sort_values("Total_Outstanding", ascending=False)
for _, row in high_risk.iterrows():
    print(f"    {row['Customer Name']:<28} {row['Governorate']:<15} EGP {row['Total_Outstanding']:>12,.2f} {row['Avg_Days_Overdue']:>12.0f} {row['Collection_Rate_%']:>11.2f}%")

# Medium risk
print(f"\n\n   MEDIUM RISK CUSTOMERS:")
print(f"  {'Customer':<30} {'Gov':<15} {'Outstanding':>15} {'Days Overdue':>12} {'Collection%':>12}")
print(f"  {'-'*85}")
med_risk = debt_analysis[debt_analysis["Risk Level"] == "Medium"].sort_values("Total_Outstanding", ascending=False)
for _, row in med_risk.iterrows():
    print(f"    {row['Customer Name']:<28} {row['Governorate']:<15} EGP {row['Total_Outstanding']:>12,.2f} {row['Avg_Days_Overdue']:>12.0f} {row['Collection_Rate_%']:>11.2f}%")

print("\n" + "=" * 75)


```

```python
#  then i want to know the Opportunities of how we can grow 


# Growth Opportunities Analysis
# Monthly trend to see which months perform best
monthly_trend = sales.groupby(["Year", "Month", "Month Name"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

monthly_trend = monthly_trend.sort_values(["Year", "Month"])

print("=" * 65)
print("   PHARMALENS — MONTHLY SALES TREND (2023 vs 2024)")
print("=" * 65)
for year in [2023, 2024]:
    print(f"\n  📅 {year}")
    year_data = monthly_trend[monthly_trend["Year"] == year]
    print(f"  {'Month':<15} {'Revenue':>15} {'Profit':>15} {'Quantity':>10}")
    print(f"  {'-'*55}")
    for _, row in year_data.iterrows():
        print(f"  {row['Month Name']:<15} EGP {row['Total_Revenue']:>12,.2f} EGP {row['Total_Profit']:>12,.2f} {int(row['Total_Quantity']):>10,}")
print("\n" + "=" * 65)

```

```python
# the forecast for the next year 


# Phase 5 - Forecast 2025
# Based on 2023 to 2024 growth rate we predict 2025

print("=" * 65)
print("   PHARMALENS — 2025 REVENUE FORECAST")
print("=" * 65)

# Calculate growth rate per governorate
gov_yearly = sales.groupby(["Governorate", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Profit=("Gross Profit EGP", "sum")
).reset_index()

print("\n   GOVERNORATE FORECAST:")
print(f"\n  {'Governorate':<20} {'2023':>15} {'2024':>15} {'Growth%':>10} {'2025 Forecast':>15}")
print(f"  {'-'*75}")

for gov in gov_yearly["Governorate"].unique():
    gov_data = gov_yearly[gov_yearly["Governorate"] == gov]
    rev_2023 = gov_data[gov_data["Year"]==2023]["Total_Revenue"].values[0]
    rev_2024 = gov_data[gov_data["Year"]==2024]["Total_Revenue"].values[0]
    growth = (rev_2024 - rev_2023) / rev_2023 * 100
    forecast_2025 = rev_2024 * (1 + growth/100)
    print(f"  {gov:<20} EGP {rev_2023:>12,.2f} EGP {rev_2024:>12,.2f} {growth:>9.2f}% EGP {forecast_2025:>12,.2f}")

# Medicine forecast
print(f"\n\n   MEDICINE FORECAST:")
med_yearly = sales.groupby(["Medicine Name", "Year"]).agg(
    Total_Revenue=("Total Revenue EGP", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

print(f"\n  {'Medicine':<25} {'2023 Rev':>15} {'2024 Rev':>15} {'Growth%':>10} {'2025 Forecast':>15}")
print(f"  {'-'*80}")

for med in med_yearly["Medicine Name"].unique():
    med_data = med_yearly[med_yearly["Medicine Name"] == med]
    rev_2023 = med_data[med_data["Year"]==2023]["Total_Revenue"].values[0]
    rev_2024 = med_data[med_data["Year"]==2024]["Total_Revenue"].values[0]
    growth = (rev_2024 - rev_2023) / rev_2023 * 100
    forecast_2025 = rev_2024 * (1 + growth/100)
    print(f"  {med:<25} EGP {rev_2023:>12,.2f} EGP {rev_2024:>12,.2f} {growth:>9.2f}% EGP {forecast_2025:>12,.2f}")

# Overall forecast
rev_2023_total = sales[sales["Year"]==2023]["Total Revenue EGP"].sum()
rev_2024_total = sales[sales["Year"]==2024]["Total Revenue EGP"].sum()
overall_growth = (rev_2024_total - rev_2023_total) / rev_2023_total * 100
forecast_2025_total = rev_2024_total * (1 + overall_growth/100)
exp_2024_total = expenses[expenses["Year"]==2024]["Actual EGP"].sum()
exp_growth = 0.10
forecast_exp_2025 = exp_2024_total * (1 + exp_growth)
forecast_profit_2025 = forecast_2025_total * 0.4323
forecast_net_2025 = forecast_profit_2025 - forecast_exp_2025

print(f"\n\n  OVERALL 2025 FORECAST:")
print(f"  {'='*65}")
print(f"  Expected Revenue       : EGP {forecast_2025_total:,.2f}")
print(f"  Expected Gross Profit  : EGP {forecast_profit_2025:,.2f}")
print(f"  Expected Expenses      : EGP {forecast_exp_2025:,.2f}")
print(f"  Expected Net Profit    : EGP {forecast_net_2025:,.2f}")
print(f"  Overall Growth Rate    : {overall_growth:.2f}%")
print(f"  {'='*65}")
```
